```python
import tensorflow as tf
import keras
import numpy as np
import utils 
import os

# Selección del dataset; descomenta el que quieras usar
# Datasets en 2D
#dataset = "2_clases_simple.csv" # LO APRENDE
#dataset = "6_clases_dificil.csv" # LO APRENDE? HAY DOS CLASES MEZCLADAS
#dataset = "moons.csv" # NO LO APRENDE. CLASES MEZCLADAS
dataset = "moons_mis-scaled.csv" #NO LO APRENDE. CLASES MEZCLADAS. DEBERIA SER UN SINUSOIDAL??
#dataset = "circulos.csv" # NO LO APRENDE. LA FRONTERA DEBERIA SER UN CIRCULO. COMO??

# Dataset con más de 2 dimensiones # COMO VER SI ESTOS MODELOS ENTRENAN BIEN?
#dataset = "iris.csv"
#dataset = "diabetes.csv"

# Carga del dataset
dataset_path=os.path.join("datasets_clasificacion",dataset)
data=np.loadtxt(open(dataset_path, "rb"), delimiter=",", skiprows=1)
x,y=data[:,0:-1],data[:,-1]
# cantidad de ejemplos y dimension de entrada
n,d_in=x.shape
# calcula la cantidad de clases
classes=int(y.max()+1)

print("Información del conjunto de datos:")
print(f"Ejemplos: {n}")
print(f"Variables de entrada: {d_in}")
print(f"Cantidad de clases: {classes}")

# Normalizo las variables de entrada
for i in range(d_in):
    x[:,i]=(x[:,i]-x[:,i].mean())/x[:,i].std()
```

```python
# Creación del modelo inicial
print("Inicialización aleatoria del modelo (podes volver a correr esta celda para obtener otros resultados)")

# iterar con distintos valores y guardar resultados
# Parametros de entrenamiento
alphas=(0.01, 0.001, 0.0001)
epochs=(50, 100, 300)
batch_size=16
resultados={
    'dataset': dataset,
    'trainings': []
}

for alpha in alphas:
    for epoch in epochs:
        # Creo un modelo logístico
        modelo = keras.Sequential([
        # la activación softmax hace que la salida sean probabilidades
        keras.layers.Dense(classes,input_shape=(d_in,), activation='softmax')])
        # visualización del modelo inicial (si hay solo 2 variables de entrada)
        print(f'------------------------ inicio entrenamiento --- epochs:{epoch} y alpha:{alpha} -----------------------')
        if d_in==2:
            utils.plot_regresion_logistica2D(modelo,x,y,title=f"Modelo inicial",detail=0.05)
        
        results = {
            'acc': None, 
            'loss': None, 
            'epochs': epoch,
            'alpha': alpha
        }
        modelo.compile(
            optimizer=keras.optimizers.SGD(lr=alpha), 
            loss='sparse_categorical_crossentropy', 
            # metricas para ir calculando en cada iteracion o batch 
            # Agregamos el accuracy del modelo
            metrics=['accuracy'], 
        )

        # Entrenamiento del modelo
        history=modelo.fit(x,y,epochs=epoch,batch_size=batch_size,verbose=0)
        results['loss']=history.history['loss'][-1]
        results['acc']=history.history['acc'][-1]
        # guardo los resultados
        resultados['trainings'].append(results)
        # visualización del modelo final (si hay solo 2 variables de entrada)
        if d_in==2:
            utils.plot_regresion_logistica2D(modelo,x,y,title=f"Modelo Final",detail=0.05)
        print(f'--------------------------------------- FIN ENTRENAMIENTO --------------------------------')
        
print("FIN")
```
